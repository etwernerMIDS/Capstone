```python
df_multiple_trades = df_trades[df_trades.groupby('Unique_id_collection')['Unique_id_collection'].transform('size') == 5420]
df_multiple_trades['Unique_id_collection'].value_counts().describe()
```

```python
# limit to relevant columns
df_multiple_trades_relevant_columns = df_multiple_trades[['Price_USD', 'Collection_cleaned', 'Seller_address', 'Buyer_address','Datetime_updated_seconds','Category']]
df_multiple_trades_relevant_columns = df_multiple_trades_relevant_columns.sort_values('Datetime_updated_seconds')
df_multiple_trades_relevant_columns
```

```python
training_set = df_multiple_trades_relevant_columns.iloc[:4000, 0:1].values
test_set = df_multiple_trades_relevant_columns.iloc[4000:, 0:1].values
print('Training set size', len(training_set), 'Test set size', len(test_set))
training_set
```

```python

X_train = []
y_train = []
for i in range(1, 4000):
    X_train.append(training_set[i-1:i, 0])
    y_train.append(training_set[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
#(740, 60, 1)
```

```python
X_train
```

```python
model = Sequential()
#Adding the first LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
model.add(Dropout(0.2))
# Adding a second LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))
# Adding a third LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))
# Adding a fourth LSTM layer and some Dropout regularisation
model.add(LSTM(units = 50))
model.add(Dropout(0.2))
# Adding the output layer
model.add(Dense(units = 1))

# Compiling the RNN
model.compile(optimizer = 'adam', loss = 'mean_squared_error')

# Fitting the RNN to the Training set
model.fit(X_train, y_train, epochs = 100, batch_size = 32)
```
