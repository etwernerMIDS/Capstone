```python
# find Google Trends weekly values
from pytrends.request import TrendReq
pytrend = TrendReq()

googletrends_dict = {}

counter = 0

for val in range(5850, 5940):
    counter = val
    pytrend.build_payload(kw_list=[collection_raw_list2[val]],timeframe='2017-07-01 2021-04-27')
    temp_df_data = pytrend.interest_over_time().reset_index()
    if len(temp_df_data) == 0:
        continue
        # googletrends_dict[item] = np.nan
    else:
        googletrends_dict[collection_raw_list2[val]] = {'date': temp_df_data.iloc[:, 0].to_list(), 'values': temp_df_data.iloc[:, 1].to_list()}
```
