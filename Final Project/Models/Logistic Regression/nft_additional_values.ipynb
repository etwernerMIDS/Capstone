# Merge social and crypto pricing data with original dataset

```python
import os
import pandas as pd
import numpy as np
os.getcwd()
```

```python
df = pd.read_csv('https://storage.googleapis.com/nft_capstone_datafiles/Data_API.csv.gz',
                dtype={'Smart_contract':str, 'ID_token':str, 'Transaction_hash':str,'Seller_username':str,'Buyer_username':str,'Image_url_2':str,
                        'Image_url_3':str,'Image_url_4':str,'Name':str,'Description':str,'Permanent_link':str})
print(len(df))
print(df.columns)
df.head(5)
```

```python
df['Datetime_updated_dt'] = pd.to_datetime(df['Datetime_updated'])
df.head(5)
```

```python
df['Datetime_daybeforesale'] = df['Datetime_updated_dt'] - pd.Timedelta(days=1)
df.head(5)
```

```python
df['nft_sales_count'] = df.groupby(['Unique_id_collection'])['Unique_id_collection'].transform('count')
# print(df.columns)
df.head(5)
```

```python
df_sorted = df.sort_values(['Category', 'Collection_cleaned', 'Datetime_updated_seconds'])
df_sorted['First_item_Collection_cleaned'] = 0
df_sorted.loc[df_sorted['Collection_cleaned'] != df_sorted['Collection_cleaned'].shift(1), 'First_item_Collection_cleaned'] = 1
```

```python
df_sorted.head(5)
```

```python
df_sorted = df_sorted.sort_values(['Category', 'Collection_cleaned', 'Unique_id_collection', 'Datetime_updated_seconds'])
df_sorted['First_item_Unique_id_collection'] = 0
df_sorted.loc[df_sorted['Unique_id_collection'] != df_sorted['Unique_id_collection'].shift(1), 'First_item_Unique_id_collection'] = 1
df_sorted.head(5)
```

```python
df_sorted['Previous_Price_USD'] = df_sorted['Price_USD'].shift(1)
df_sorted['Previous_Datetime_updated_seconds'] = df_sorted['Datetime_updated_seconds'].shift(1)
df_sorted['Elapsed_time'] = np.where(df_sorted['First_item_Unique_id_collection'] == 1, pd.NaT, 
                                     pd.to_datetime(df_sorted['Datetime_updated_seconds']) - pd.to_datetime(df_sorted['Previous_Datetime_updated_seconds']))
df_sorted['Price_increase'] = np.where(df_sorted['First_item_Unique_id_collection'] == 1, 0, 
                                       np.where(df_sorted['Price_USD']-df_sorted['Previous_Price_USD']>0,1,0))
df_sorted['Price_diff'] = np.where(df_sorted['First_item_Unique_id_collection'] == 1, np.NaN, df_sorted['Price_USD']-df_sorted['Previous_Price_USD'])
df_sorted['Price_diff_pctg'] = np.where(df_sorted['First_item_Unique_id_collection'] == 1, np.NaN, 
                                        (df_sorted['Price_USD']/df_sorted['Previous_Price_USD'])-1)
df_sorted.head(5)

# df_classification2['Price_increase'] = np.where(df_classification2['First_item_Unique_id_collection'] == 1, 0, df_classification2['Price_increase'])
# df_classification2['Price_diff'] = np.where(df_classification2['First_item_Unique_id_collection'] == 1, np.NaN, df_classification2['Price_diff'])
# df_classification2['Price_diff_pctg'] = np.where(df_classification2['First_item_Unique_id_collection'] == 1, np.NaN, df_classification2['Price_diff_pctg'])
# df_classification2['Elapsed_time'] = np.where(df_classification2['First_item_Unique_id_collection'] == 1, pd.NaT, df_classification2['Elapsed_time'])
```

```python
# Create smaller dataframe with
# Collection_cleaned
# Datetime_updated
# Price_USD
# Price_Crypto

df_sorted5_priceprep = df_sorted5[['Collection_cleaned', 'Datetime_updated', 'Price_USD', 'Price_Crypto']]
print(len(df_sorted5_priceprep))
df_sorted5_priceprep.head(5)
```

```python
# col_list = ['Price_USD', 'Price_Crypto']
# window_list = [7, 14, 30, 60, 90, 180, 365, 730]
# agg_list = ['median', 'max']

# Write for loop to create columns for all new variations
# Delete rows for all but the last row for each Collection_cleaned/Datetime_updated combination
# Create new column that adds 1 day to the Datetime_updated column
# Join new dataframe with dummy dataframe that includes 1 row per day per Collection_cleaned - may not need to do this - problem is that the dates may not align
# Alternative
# Join new dataframe with df_sorted5 on 'Collection_cleaned' and  new date column with df_sorted5['Datetime_daybeforesale']

col_list = ['Price_USD', 'Price_Crypto']
window_list = [7, 14, 30, 60, 90, 180, 365, 730]
agg_list = ['median', 'max']
```

```python
%%time
# Write for loop to create columns for all new variations

for val in col_list:
    for item in window_list:
        for func in agg_list:
            if func == 'median':
                df_sorted5_priceprep[val+'_'+func+'_'+str(item)+'d'] = df_sorted5_priceprep.groupby(['Collection_cleaned', 'Datetime_updated'])[val].transform(lambda s: s.rolling(item, min_periods=1).median())
            else:
                df_sorted5_priceprep[val+'_'+func+'_'+str(item)+'d'] = df_sorted5_priceprep.groupby(['Collection_cleaned', 'Datetime_updated'])[val].transform(lambda s: s.rolling(item, min_periods=1).max())
    
    # df_collections_otherdata3[str(item)+'d_rolling_avg'] = df_collections_otherdata3.groupby('Collection').rolling(item)['values'].mean().reset_index(drop=True)
    # df_collections_otherdata3[str(item)+'d_rolling_avg'] = df_collections_otherdata3.groupby('Collection')['values'].transform(lambda s: s.rolling(item, min_periods=1).mean())
    #test_df3['Ggl_trends_collection_'+str(item)+'d_rolling_avg'] = test_df3.groupby('Collection')['Ggl_trends_collection'].transform(lambda s: s.rolling(item, min_periods=1).mean())
```
