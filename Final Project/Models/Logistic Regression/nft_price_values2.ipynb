## Calculate median price of primary and secondary sales made in the collection of interest during a time window prior to t_s

```python
# Load data
df = pd.read_csv('/home/jupyter/nft/Data_API.csv.gz',
                 dtype={'Smart_contract':str, 'ID_token':str, 'Transaction_hash':str,'Seller_username':str,'Buyer_username':str,'Image_url_2':str,
                        'Image_url_3':str,'Image_url_4':str,'Name':str,'Description':str,'Permanent_link':str})
```

```python
df.columns
```

```python
df1 = df.sort_values(['Collection_cleaned', 'Datetime_updated_seconds', 'Price_USD'], ascending=True)
df1.head(10)
```

```python
df_collection_primary_sales = df1.groupby('Collection_cleaned', as_index=False)['Datetime_updated_seconds','Price_USD','Unique_id_collection'].first()
print(len(df_collection_primary_sales))
df_collection_primary_sales.head(5)
```

```python
df2 = df.sort_values(['Unique_id_collection', 'Datetime_updated_seconds', 'Price_USD'], ascending=True)
df2.head(10)
```

```python
df_unique_nft_primary_sales = df2.groupby('Unique_id_collection', as_index=False)['Datetime_updated_seconds','Price_USD'].first()
print(len(df_unique_nft_primary_sales))
df_unique_nft_primary_sales.head(5)
```

```python
df.Unique_id_collection.nunique()
```

```python
df1["Median_USD_all"] = df1.groupby(['Collection_cleaned','Datetime_updated'])['Price_USD'].expanding().median().values
df1["Median_crypto_all"] = df1.groupby(['Collection_cleaned','Datetime_updated'])['Price_Crypto'].expanding().median().values 
df1.head(10)
```

```python
df3 = df1.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_USD_2y = d['Price_USD'].rolling(window = 730, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_Crypto_2y = d['Price_Crypto'].rolling(window = 730, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_USD_1y = d['Price_USD'].rolling(window = 365, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_Crypto_1y = d['Price_Crypto'].rolling(window = 365, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_USD_180d = d['Price_USD'].rolling(window = 180, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_Crypto_180d = d['Price_Crypto'].rolling(window = 180, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_USD_90d = d['Price_USD'].rolling(window = 90, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_Crypto_90d = d['Price_Crypto'].rolling(window = 90, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_USD_60d = d['Price_USD'].rolling(window = 60, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_Crypto_60d = d['Price_Crypto'].rolling(window = 60, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_USD_30d = d['Price_USD'].rolling(window = 30, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_Crypto_30d = d['Price_Crypto'].rolling(window = 30, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_USD_14d = d['Price_USD'].rolling(window = 14, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_Crypto_14d = d['Price_Crypto'].rolling(window = 14, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_USD_7d = d['Price_USD'].rolling(window = 7, min_periods = 1).median())).reset_index(drop = True)
df3 = df3.groupby(['Collection_cleaned']).apply(lambda d: d.assign(Median_Crypto_7d = d['Price_Crypto'].rolling(window = 7, min_periods = 1).median())).reset_index(drop = True)
```
